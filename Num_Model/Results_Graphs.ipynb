```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('results.csv')
data_WCIT = pd.read_csv('results_WCIT.csv')
```

```python
fig, axs = plt.subplots(1, 2)
fig.set_figheight(4)
fig.set_figwidth(10)

axs[0].plot(data['1'], data['2'], color=(0.0, 0.0, 0.0))
axs[0].plot(data['5'], data['6'], color=(0.5, 0.5, 0.5))
axs[0].plot(data['9'], data['10'], color=(0.8, 0.8, 0.8))
axs[0].plot(data_WCIT['1'], data_WCIT['2'], color=(0.0, 0.0, 0.0), linestyle='--')
axs[0].plot(data_WCIT['5'], data_WCIT['6'], color=(0.5, 0.5, 0.5), linestyle='--')
axs[0].plot(data_WCIT['9'], data_WCIT['10'], color=(0.8, 0.8, 0.8), linestyle='--')
#axs[0].set_title('')
axs[0].set(xlabel='Young generation political power ($\mu$)',ylabel='Market interest rate per year in %')
axs[0].set_xlim([0.25, 2.0]);
axs[0].set_ylim([4.0, 6.0]);

axs[1].plot(data['1'], data['3'], color=(0.0, 0.0, 0.0))
axs[1].plot(data['5'], data['7'], color=(0.5, 0.5, 0.5))
axs[1].plot(data['9'], data['11'], color=(0.8, 0.8, 0.8))
axs[1].plot(data_WCIT['1'], data_WCIT['3'], color=(0.0, 0.0, 0.0), linestyle='--')
axs[1].plot(data_WCIT['5'], data_WCIT['7'], color=(0.5, 0.5, 0.5), linestyle='--')
axs[1].plot(data_WCIT['9'], data_WCIT['11'], color=(0.8, 0.8, 0.8), linestyle='--')
#axs[1].set_title('')
axs[1].set(xlabel='Young generation political power ($\mu$)',ylabel='Emission interest rate per year in %')
axs[1].set_xlim([0.25, 2.0]);
axs[1].set_ylim([4.0, 6.0]);
```
